```python
#To make the appearance of our game look totally awesome we're going to use the IPython.display module
#Now let's use the import display and Image classes from the module
#This method is super common to display images across Python engineering environments


from IPython.display import display, Image
#Now we're going to utilize another module which is the random module in Python
#The random module is going to have the computer choose randonly from our 3 choices
import random

#We now are going to insert our image files we can do this by creating an assignment statement
#This is our variable (menu_img)
menu_img = Image(filename='combat.png')  #This argument shows our image for the menu because of the display function we used in the first line of code
rock_img = Image(filename='rock.jpg')
paper_img = Image(filename='paper.png')  
scissors_img = Image(filename='scissors.jpg')
quit_img = Image(filename='quit.png') 

#Now We're Going to create our main page which is our menu of playable otions for our gamer
print("READY FOR HOLY BATTLE!?")
display(menu_img)#This simple function here makes the magical image appear (cool)

#We're going to create our loop with a while True
while True:
    options = ["ROCK", "PAPER", "SCISSORS"]
    computer_choice = random.choice(options)#By using random.choice we now have the computer selecting randomly from rock, paper and scissors from our options list
    #random.choice is our function (.choice specifically) from our random module
    player_choice = input("ENTER YOUR CHOICE! (ROCK, PAPER, SCISSORS), OR TYPE 'I QUIT' TO END THE GAME ") #Create our user's input (function)
    if player_choice.upper() == "I QUIT":
        display(quit_img)
        break#Cool simple way to go back and set code to caps is using the .upper() method
#And we're going to break the loop when the gamer decides to choose I QUIT.
    if player_choice == "ROCK": 
        player_img = rock_img
    elif player_choice == "PAPER":
        player_img = paper_img
    else:
        player_img = scissors_img
    #Through utilizing our if, elif and else statements we're creating the parameters for our gane
    if computer_choice == "ROCK":
        computer_img = rock_img
    elif computer_choice == "PAPER":
        computer_img = paper_img
    else:
        computer_img = scissors_img
#We do the same but making sure all options and choices are exhausted correctly abd displays the correct outcomes
    if player_choice == computer_choice:
        result = "GAME TIED"
    elif player_choice == "ROCK":
        result = "YOU WIN!" if computer_choice == "SCISSORS" else "YOU LOSE"
    elif player_choice == "PAPER":
        result = "YOU WIN!" if computer_choice == "ROCK" else "YOU LOSE"
    elif player_choice == "SCISSORS":
        result = "YOU WIN!" if computer_choice == "PAPER" else "YOU LOSE"

    #Utlizing display we're going to cause the game to display our results from the previos round of set of rounds.
    display(player_img, computer_img) 
    print(result)

print("GREAT BATTLE! I WILL RETURN FOR REVENGE!")

#This was a lot of fun creating this game and I totally got these images and idea from the Rock, Paper, Scissors anime -https://youtu.be/GVT3WUa-48Y?si=qpCuS-idBG7QOPO7
#Also found a really cool free photoshop alternative and customized a wee bit @PhotoMind app on Microsoft Store 100% free though! :)
```
